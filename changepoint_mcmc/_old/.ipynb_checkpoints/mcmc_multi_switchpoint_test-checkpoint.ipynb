```python
nrn = 6
this_dat = dat.spikes[taste][:,nrn]
print(this_dat.shape)
```

```python
spike_inds = np.where(this_dat)
plt.scatter(spike_inds[1],spike_inds[0],alpha = 0.2)
```

```python
# Convert spikes to counts by binning
time_lims = [2000,6000]
bin_width = 50
t_vec = np.arange(this_dat.shape[-1])
binned_t_vec = np.mean(t_vec[time_lims[0]:time_lims[1]].reshape((-1,bin_width)),axis=-1)
this_dat_binned = np.sum(this_dat[...,time_lims[0]:time_lims[1]].reshape(this_dat.shape[0],-1,bin_width),axis=-1)
this_dat_binned = np.vectorize(np.int)(this_dat_binned)
plt.plot(binned_t_vec,this_dat_binned.T,alpha =0.5);plt.show()
```

```python
#trial = 0
#count_data = this_dat_binned[trial]
count_data = np.sum(this_dat_binned,axis=0)
n_count_data = len(count_data)
print(count_data)
print('Average count = {}'.format(count_data.mean()))
plt.bar(np.arange(n_count_data),count_data,color="#348ABD");plt.show()
```

```python
states = 4
with pm.Model() as model:
    idx = np.arange(n_count_data) # Index
    # Defining variables with priors
    alpha = 1.0/count_data.mean()  # Recall count_data is the
                                   # variable that holds our txt counts
    #lambda_1 = pm.Exponential("lambda_1", alpha)
    #lambda_2 = pm.Exponential("lambda_2", alpha)
    #lambda_3 = pm.Exponential("lambda_3", alpha)
    
    #lambda_latent = pm.math.stack([pm.Exponential("lambda_{}".format(i+1),alpha) for i in range(states)])
    lambda_latent = pm.Exponential("lambda_latent", alpha, shape = states) 
    
    #tau_1 = pm.DiscreteUniform("tau_1", lower=0, upper=n_count_data)
    #tau_2 = pm.DiscreteUniform("tau_2", lower=tau_1, upper=n_count_data)
    
    # Scaled beta mixture "somewhere in the middle" prior
    a = pm.HalfNormal('a_hp', 3.)
    b = pm.HalfNormal('b_hp', 3.)
    
    #tau_latent = pm.math.stack([pm.Beta("tau_{}_latent".format(i+1),a,b) for i in range(states-1)])
    #tau_1 = pm.Deterministic('tau_1', idx.min() + (idx.max() - idx.min()) * tau_1_latent)
    #tau_2 = pm.Deterministic('tau_2', tau_1 + (idx.max() - idx.min()) * tau_2_latent)
    
    #tau_latent = pm.Dirichlet('tau',np.array([1,3]))
    #tau_1 = pm.Deterministic('tau_1', (idx.max() - idx.min() * tau_latent[0]))
    #tau_2 = pm.Deterministic('tau_2', (idx.max() - idx.min() * tau_latent[1]))
    
    
    tau_names = ['tau_{}'.format(state+1) for state in range(states-1)]
    tau_latent = pm.Beta("tau_latent",a,b, shape = states -1)
    tau_list = []
    for state in range(states-1):
        if state == 0:
            tau_list.append(
                pm.Deterministic(tau_names[state], idx.min() + (idx.max() - idx.min()) * tau_latent[state])
            )
        elif state = (states-1):
            tau_list.append(
                pm.Deterministic(tau_names[state], tau_list[state-1] + (idx.max() - idx.min()) * tau_latent[state])
            )
        else:
            tau_list.append(
                pm.Deterministic(tau_names[state], tau_list[state-1] + (idx.max() - idx.min()) * tau_latent[state])
            )
    tau_list = pm.math.stack(tau_list)
    
    #tau_1 = pm.Deterministic('tau_1', idx.min() + (idx.max() - idx.min()) * tau_latent[0])
    #tau_2 = pm.Deterministic('tau_2', tau_1 + (idx.max() - idx.min()) * tau_latent[1])
    

    
    sigmoid_list = pm.math.stack([tt.nnet.sigmoid(2 * (idx - tau_list[i])) for i in range(states-1)])
    flipped_sigmoid_list = 1 - sigmoid_list
    
    lambda_list = []
    for state in range(states):
        if state == 0:
            lambda_list.append(
                flipped_sigmoid_list[state] * lambda_latent[state]
            )
        elif state == states-1:
            lambda_list.append(
                sigmoid_list[state-1] * lambda_latent[state]
            )
        else:
            lambda_list.append(
                sigmoid_list[state-1] * flipped_sigmoid_list[state] * lambda_latent[state]
            )
    lambda_ = pm.math.sum(lambda_list,axis=0)
    
    #lambda_ = flipped_sigmoid_list[0] * lambda_latent[0] + \
    #        sigmoid_list[0] * flipped_sigmoid_list[1] * lambda_latent[1] + \
    #        sigmoid_list[1] * lambda_latent[2]
    
    #weight_1 = tt.nnet.sigmoid(2 * (idx - tau_list[0]))
    #weight_2 = tt.nnet.sigmoid(2 * (idx - tau_list[1]))
    #lambda_ = (1 - weight_1) * lambda_latent[0] + \
    #            weight_1 * (1 - weight_2) * lambda_latent[1] + \
    #            weight_2 * lambda_latent[2]
#print(lambda_.tag.test_value)
    
    
with model:
    observation = pm.Poisson("obs", lambda_, observed=count_data)
```

```python
with model:
    #step = pm.Metropolis()
    #trace = pm.sample(15000, tune=5000,step=step)
    trace = pm.sample(50000, tune=25000, chains = 8, cores = 8)
```
