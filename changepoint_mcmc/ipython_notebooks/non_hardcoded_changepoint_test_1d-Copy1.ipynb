```python
# x axis
x = np.arange(100)

# locations on axis of the changepoints
cps = np.array([20, 40])

# helper function
def cp_design_mat(x, cps):
    return (0.5 * (1.0 + np.sign(np.tile(x[:,None], (1,len(cps))) - cps)))

A = cp_design_mat(x, cps)

plt.plot(x,A)
```

```python
# the following are unknown parameters that you can estimate by giving 
# them priors and then sampling with NUTS
k = 0.1  # initial slope
m = 0.3 # offset
delta = np.array([0.7, -1]) # slope parameters

# generate the function at x values
y = (k + np.dot(A, delta))*x + (m + np.dot(A, -cps * delta))

plt.plot(x, y)
plt.plot(x,k*x + m, alpha = 0.5)
plt.plot(x+cps[0],(k+delta[0])*x + m, alpha = 0.5)
plt.plot(x+cps[1],(k+sum(delta))*x + m, alpha = 0.5)
```

```python
#data_vec = np.random.random(len(x)) < y
data_vec = np.random.poisson(y)

fig,ax = plt.subplots(3,1)
ax[0].plot(y)
#ax[0].set_ylim(0,1)
ax[1].plot(data_vec)
ax[2].plot(np.convolve(data_vec,np.ones(20)))
ax[0].set_ylabel("True Rate")
ax[1].set_ylabel("Spikes")
ax[2].set_ylabel("Firing Rate")
```
